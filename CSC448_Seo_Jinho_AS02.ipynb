```python
# import libraries
import numpy as np
import random
import math 
import time
```

```python
def naiv_sum(N):
    ''' Return Naive Sum'''
    sum = 0.0
    
    for i in N:
        sum += i

    return sum

def comp_sum(N):
    ''' Return Conpensated Sum'''
    sum = 0.0
    c = 0.0
    
    for i in N:
        y = i - c
        temp = sum + y
        c = (temp - sum) - y
        sum = temp
        
    return sum


npows = [pow(10, 5), pow(10, 6), pow(10, 7)]
for n in npows:
    # generate array of numbers

    x = np.zeros(n)
    
    id = 14225432         # use your student id number here.
    random.seed(id)
    id10 = id/10**10
        
    for i in range(n):
        x[i] = random.uniform(0.,2.) + id10
    
    print("\n")
    # Assume that math.fsum returns exact value(s)
    print("N:", n)
    print("{:<17}{}".format("Exact Value:", math.fsum(x)))
    print("---------------------------------------")
    
    # Naive Sum
    start_time = time.time()
    print("{:<17}{}".format("Naive Sum:", naiv_sum(x)))
    end_time = time.time()
    print("{:<17}{}".format("Excution Time:", end_time-start_time))

    # Naive Sum Errors 
    print("{:<17}{}".format("Absolute Error:", abs(naiv_sum(x) - math.fsum(x))))
    print("{:<17}{}".format("Relative Error:", abs((naiv_sum(x) - math.fsum(x)) / math.fsum(x))))
    print("---------------------------------------")
    
    # Numpy Sum
    start_time = time.time()
    print("{:<17}{}".format("Numpy Sum:", np.sum(x)))
    end_time = time.time()
    print("{:<17}{}".format("Excution Time:", end_time-start_time))

    # Numpy Sum Error 
    print("{:<17}{}".format("Absolute Error:", abs(np.sum(x) - math.fsum(x))))
    print("{:<17}{}".format("Relative Error:", abs((np.sum(x) - math.fsum(x)) / math.fsum(x))))
    print("---------------------------------------")
    
    # Compensated Sum
    start_time = time.time()
    print("{:<17}{}".format("Compensated Sum:", comp_sum(x)))
    end_time = time.time()
    print("{:<17}{}".format("Excution Time:", end_time-start_time))

    # Compensated Sum Error 
    print("{:<17}{}".format("Absolute Error:", abs(comp_sum(x) - math.fsum(x))))
    print("{:<17}{}".format("Relative Error:", abs((comp_sum(x) - math.fsum(x)) / math.fsum(x))))
    
    print("_______________________________________")


```
